### Data Wrangling Code

**Playstore Dataset Data Wrangling**

````python
# Write your code to make your dataset analysis ready.
````

````python
# changing a wrong entry into obvious entry

for i in range(10841):
  if playstore_df.loc[i,'Reviews']=='3.0M':
    playstore_df.loc[i,'App']='Life Made WI-Fi Touchscreen Photo Frame'
    playstore_df.loc[i,'Category']=''
    playstore_df.loc[i,'Rating']=1.9
    playstore_df.loc[i,'Reviews']='19'
    playstore_df.loc[i,'Size']='3.9M'
    playstore_df.loc[i,'Installs']='1,000+'
    playstore_df.loc[i,'Type']='Free'
    playstore_df.loc[i,'Price']=0
    playstore_df.loc[i,'Content Rating']='Everyone'
    playstore_df.loc[i,'Genres']=''
    playstore_df.loc[i,'Last Updated']='February 11, 2018'
    playstore_df.loc[i,'Current Ver']='1.0.19'
    playstore_df.loc[i,'Android Ver']='4.0 and up'
````

````python
# other than rating, we will convert reviews-int,size-float,installs-int, price-float

#Reviews--to be converted to int
playstore_df['Reviews']= playstore_df['Reviews'].astype('int')

for i in range(10841):

  #Size--to be converted into kb  by removing mb and kb keywords
  if playstore_df.loc[i,'Size'][-1]=='M':
    playstore_df.loc[i,'Size']=float(playstore_df.loc[i,'Size'].replace('M',''))*1000

  elif playstore_df.loc[i,'Size'][-1]=='k':
    playstore_df.loc[i,'Size']=float(playstore_df.loc[i,'Size'].replace('k',''))
  
  else:
    playstore_df.loc[i,'Size']=-1    #-1 to be allocated to those apps whose size varies with device

  #Installs--to be converted to integer by removing + and ,
  if playstore_df.loc[i,'Installs'][-1]=='+':
    playstore_df.loc[i,'Installs']=int((playstore_df.loc[i,'Installs'].replace('+','')).replace(',',''))
  else:
    playstore_df.loc[i,'Installs']=int(playstore_df.loc[i,'Installs'])  

  #price--to be converted to float by removing $ 
  try:
    if playstore_df.loc[i,'Price'][0]=='$':
      playstore_df.loc[i,'Price']=float(playstore_df.loc[i,'Price'].replace('$',''))
  except:
    playstore_df.loc[i,'Price']=float(playstore_df.loc[i,'Price'])


playstore_df['Size']=playstore_df['Size'].astype('int')
playstore_df['Installs']=playstore_df['Installs'].astype('int')
playstore_df['Price']=playstore_df['Price'].astype('int')
````

````python
# renaming column name from size to size in kb
playstore_df.rename(columns={'Size':'Size in kb'},inplace=True)
````

````python
range1=range(0,1000000)
range2=range(1000000,10000000)
range3=range(10000000,100000000)
range4=range(100000000,500000000)
range5=range(500000000,1000000000)
range6=range(1000000000,)
````

````python
def custom_agg(x):
  if x in range1:
    return '0-10 Lac'
  elif x in range2:
    return '10 Lac-1 Cr'
  elif x in range3:
    return '1 Cr-10 Cr'
  elif x in range4:
    return '10 Cr-50 Cr'
  elif x in range5:
    return '50 Cr-100 Cr'
  else:
    return '100 Cr+'
````

````python
communication_df=playstore_df[playstore_df['Category']=='COMMUNICATION'].groupby('Installs').agg({"App":'count'}).sort_values('App',ascending=False)

communication_df['Install range']=communication_df.index.map(custom_agg)
communication_df=communication_df.groupby('Install range').agg({'App':'sum'})
communication_df
````

````python
game_df=playstore_df[playstore_df['Category']=='GAME'].groupby('Installs').agg({"App":'count'}).sort_values('App',ascending=False)

game_df['Install range']=game_df.index.map(custom_agg)
game_df=game_df.groupby('Install range').agg({'App':'sum'})
game_df
````

````python
productivity_df=playstore_df[playstore_df['Category']=='PRODUCTIVITY'].groupby('Installs').agg({"App":'count'}).sort_values('App',ascending=False)

productivity_df['Install range']=productivity_df.index.map(custom_agg)
productivity_df=productivity_df.groupby('Install range').agg({'App':'sum'})
productivity_df
````

````python
family_df=playstore_df[playstore_df['Category']=='FAMILY'].groupby('Installs').agg({"App":'count'}).sort_values('App',ascending=False)

family_df['Install range']=family_df.index.map(custom_agg)
family_df=family_df.groupby('Install range').agg({'App':'sum'})
family_df
````

````python
social_df=playstore_df[playstore_df['Category']=='SOCIAL'].groupby('Installs').agg({"App":'count'}).sort_values('App',ascending=False)

social_df['Install range']=social_df.index.map(custom_agg)
social_df=social_df.groupby('Install range').agg({'App':'sum'})
social_df
````
